** Check whether all the zones generate the same amount of sales.**

```python
import pandas as pd
data=pd.read_csv('/content/sample_data/Sales_data_zone_wise.csv')
#H0:Amount of sales is same.
#H1:Amount of sales is different.
alpha=0.05
from scipy.stats import f_oneway
f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

since pvalue is less than alpha,we reject null hypothesis.So we conclude that all the zones generate different amount of sales.
